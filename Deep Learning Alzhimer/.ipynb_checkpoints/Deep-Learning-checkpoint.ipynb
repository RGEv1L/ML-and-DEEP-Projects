```python
import matplotlib.pyplot as plt
import tensorflow
```

```python


from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, BatchNormalization, Dropout, Flatten, Dense, Activation, MaxPool2D, Conv2D
from tensorflow.keras.applications import DenseNet169



from tensorflow.keras.applications import DenseNet121,InceptionV3, Xception, ResNet101
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,validation_split=0.2)

```

```python
train_dataset  = train_datagen.flow_from_directory(directory = '/home/AKhan/Downloads/DEEP LEARNING PROJECT/Alzheimer_s Dataset/train',
                                                   target_size = (75,75),
                                                   class_mode = 'categorical',
                                                   subset = 'training',
                                                   batch_size = 10)
```

```python
valid_dataset = train_datagen.flow_from_directory(directory = '/home/AKhan/Downloads/DEEP LEARNING PROJECT/Alzheimer_s Dataset/train',
                                                  target_size = (75,75),
                                                  class_mode = 'categorical',
                                                  subset = 'validation',
                                                  batch_size = 10)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('/home/AKhan/Downloads/DEEP LEARNING PROJECT/Alzheimer_s Dataset/train/MildDemented/mildDem0.jpg')
imgplot = plt.imshow(img)
plt.show()
```

```python
base_model = DenseNet121(input_shape=(75,75,3), 
                         include_top=False)
```

```python
del model
```

```python
for layer in base_model.layers:
    layer.trainable=False
```

```python
model=Sequential()
model.add(base_model)
model.add(Flatten())
model.add(BatchNormalization())
model.add(Dense(512,activation='relu'))
model.add(Dense(4,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              metrics=['accuracy'],
              optimizer=tensorflow.keras.optimizers.Adam(lr=0.001))
```

```python
model_history=model.fit(train_dataset,
                        validation_data=valid_dataset,
                        epochs=30,batch_size=50)
```

```python
#Plotting training and validation accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
